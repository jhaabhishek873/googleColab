Label Encoding 


*   Converting the Labels into numeric form





```python
#importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to Pandas datFrame
cancer_data = pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
#finding the count of Diffrent Labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the Label Encounter Function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to the DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

Label Encoding of Iris Data

```python
# loading the data from csv file to Pandas datFrame
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
iris_data.head()
```

```python
label_encode = LabelEncoder()
```

```python
iris_label = label_encode.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_label
```
